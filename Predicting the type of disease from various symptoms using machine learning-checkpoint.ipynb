## Extracting independent(features) and dependent(lables) variables

```python
X_train = dftrain1.drop('prognosis', axis=1)
y_train = dftrain1['prognosis']
y_train = np.array(y_train).reshape(y_train.shape[0], 1)

X_test = dftest.drop('prognosis', axis=1)
y_test = dftest['prognosis']
y_test = np.array(y_test).reshape(y_test.shape[0], 1)

print("X_train : {} \ny_train : {} \nX_test: {} \ny_test : {}".format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))
```

```python
plt.figure(figsize = (30, 30))
sns.heatmap(X_train.corr(), cmap = 'PuBu', annot = False)
plt.show()
```

```python
def get_corr(data, threshold):
    corr_col = set() # Set of names of all correlated columns
    cormat = data.corr()
    for i in range(len(cormat.columns)):
        for j in range(i):
            if abs(cormat.iloc[i, j])>threshold:
                col_name = cormat.columns[i]
                corr_col.add(col_name)
    return corr_col
```

```python
corr_features = get_corr(X_train, 0.80)
len(corr_features)
```

```python
X_train_uncorr = X_train.drop(labels=corr_features, axis=1)
X_test_uncorr = X_test.drop(labels=corr_features, axis=1)
```

## 1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train_uncorr,y_train) 
```

```python
y_pred=reg.predict(X_test_uncorr)
y_pred_proba = reg.predict_proba(X_test_uncorr)
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
print("Classification Report is:\n",classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (14,11))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
accuracy_lr=round(accuracy_score(y_test,y_pred)* 100, 2)
print(accuracy_lr)
```

```python
from sklearn.metrics import f1_score
f1_score_lr=(f1_score(y_test, y_pred, average='micro'))
print(f1_score_lr)
```
